```python
import pandas as pd    # IMPORT NECESSARY MODULES
import numpy as np
```

```python
data = pd.read_csv('/content/Zomato_first_page.csv')      # READ THE CSV FILE CONTAINING THE MAIN PAGE DATA
```

```python
zomato = pd.DataFrame(data)          # CONVERT THE CSV FILE DATA TO DATFRAME FOR FURTHER PROCESSING
```

```python
zomato   # CHECK DATA
```

```python
zomato['price'] = zomato['price'].str.replace('₹', "")     # USE REPLACE TO REMOVE '₹' SIGN FROM PRICE COLUMN
```

```python
zomato['price'] = zomato['price'].str.replace(',', "")      # USE REPLACE TO REMOVE "," FROM PRICE
```

```python
zomato['price'] = zomato['price'].str.replace(' for two', "")   # USE REPLACE TO REMOVE "FOR TWO" FROM
```

```python
zomato['price'] = zomato['price'].astype(int)  # CONVERT PRICE COLUMN DATA TYPE TO INTEGER
```

```python
zomato['price'].value_counts()   # CHECK IF THE NECESSARY CHANGES ARE SUCCESFUL
```

```python
zomato['rating'].value_counts()
```

```python
zomato['rating'] = zomato['rating'].str.replace('New', "0.0")  # REMOVE 'New' FROM RATING COLUMN
zomato['rating'] = zomato['rating'].str.replace('-', "0.0")    # REMOVE '-' FROM RATING COLUMN
```

```python
zomato['rating'] = zomato['rating'].astype(float)   # CHANGE DATA TYPE OF RATING COLUMN TO FLOAT
```

```python
mean_rating = round(zomato['rating'].mean(),1)    # CALCULATE THE MEAN RATING
```

```python
print(mean_rating)  # CHECK MEAN VALUE
```

```python
zomato['rating'] = zomato['rating'].replace(0.0, mean_rating) # SINCE THE DATA HAS ONLY HANDFUL RATINGS WITH VALUE 0 REPLACE THOSE VALUES WITH MEAN
```

```python
zomato['rating'].value_counts()    # CHECK IF THE CHANGES ARE SUCCESFULL
```

```python
zomato   # FINAL CHECK
```

```python
zomato_sec = pd.read_csv('/content/zomato_second_page.csv')  # LOAD CSV FILE CONTAINING SECOND PADE DATA
```

```python
zomato_2 = pd.DataFrame(zomato_sec)    # CONVERT DATA TO DATA FRAME
```

```python
zomato_2 # CHECK DATA
```

```python
zomato['name'] = zomato['name'].str.replace('\n', "")  # REPLACE '\N' WITH NOTHING TO REMOVE
```

```python
print(zomato['name'].str.replace(' ', "/")) # REPLACE BLANK SPACE WITH '/' TO SEPERATE THE VALUES
```

```python
zomato['name'].value_counts()               # CHECK TO SEE IF DUPLICATE NAMES EXIST
```

```python
c = zomato_2['name'].value_counts()   # FIND COUNT OF OCCURENCE OF EACH HOTEL NAME
c = c[c==2].index                     # FIND NAMES WHOSE COUNT IS EQUAL 2
print(c)                              # PRINT THE NAMES WHICH OCCUR TWICE
```

```python
merged_zomato = pd.merge(zomato, zomato_2, on="name")  # MERGE THE MAIN PAGE DATA WITH SECOND PAGE DATA ON THE BASIS OF NAME
```

```python
merged_zomato # CHECK IF MERGE SUCCESFUL
```

```python
merged_zomato.drop("Unnamed: 0_x", axis=1, inplace=True)  # DROP COLUMN WITH THE NAME 'Unnamed: 0_x' FROM THE MERGED DATA
```

```python
merged_zomato # CHECK IF DROP SUCCESFUL
```

```python
merged_zomato['restaurant_id'] = range(1, len(merged_zomato) + 1)  # ADD A COLUMN IN THE DATASET WITH VALUES PROVIDED IN THE RANGE
```

```python
merged_zomato = merged_zomato.rename(columns={'name': 'restaurant_name', 'link': 'restaurant_link',}) # RENAME COLUMN NAMES
```

```python
merged_zomato.columns # CHECK THE COLUMN NAMES OF MERGED DATA
```

```python
cols = ['restaurant_id', 'restaurant_name', 'price','restaurant_link','rating','cuisine', 'location','Timings','dish_name','Resturant_known_for',
       'delivery_review_number']
merged_zomato = merged_zomato[cols]  # REAARANGE THE DATAFRAME TO INCLUDE ONLY THE SPECIFIED COLUMNS AND REMOVE ANY OTHER COLUMN NOT SPECIFIED IN THE CODE
```

```python
merged_zomato # CHECK IF SUCCESFUL
```

```python
merged_zomato['Timings'] = merged_zomato['Timings'].str.replace('(Today)', "")  # REMOVE 'Today' FROM TIMINGS COLUMN
```

```python
merged_zomato  # CHECK IF SUCCESFUL
```

```python
merged_zomato['delivery_review_number'] = merged_zomato['delivery_review_number'].str.replace(',', "") # REMOVE ',' FROM THE MENTIONED COLUMN
```

```python
merged_zomato['delivery_review_number'] = merged_zomato['delivery_review_number'].str.replace('.', "") # REMOVE '.'FROM MENTIONED COLUMN
merged_zomato['delivery_review_number'] = merged_zomato['delivery_review_number'].str.replace('K', "00") # REPLACE 'K' WITH 00
```

```python
merged_zomato['delivery_review_number'].value_counts() # CHECK IF SUCCESFULL
```

```python
merged_zomato['delivery_review_number'] = merged_zomato['delivery_review_number'].astype(int) # CHANGE THE DATA TYPE OF COLUMN TO INT
```

```python
merged_zomato # CHECK IF FUNCTION SCCESFULL
```

```python
merged_zomato['Timings'] = merged_zomato['Timings'].str.replace('midnight', "am") # REPLACE 'MIDNIGHT' WITH 'am'FROM TIMINGS COLUMN
merged_zomato['Timings'] = merged_zomato['Timings'].str.replace('noon', "pm")     # REPLACE 'NOON' WITH 'pm'FROM TIMINGS COLUMN
```

```python
merged_zomato # CHECK IF SUCCESFULL
```

```python
merged_zomato['Timings'].value_counts() # CHECK FOR ANY DISCREPANCIES
```

```python
# CHANGE DATA TYPE OF RESTURANT KNOW FOR TO STRING AND REPLACE 'Exclusive of applicable taxes and charges, if any' WITH 'Nan'
merged_zomato['Resturant_known_for'] = merged_zomato['Resturant_known_for'].str.replace("Exclusive of applicable taxes and charges, if any", "Nan")
```

```python
merged_zomato # CHECK IF SUCCESFULL
```

```python
pattern = r'₹\d{1,3}(?:,\d{3})* for two people \(approx\.\)(?: with alcohol)?'  # VARIABLE WITH PATTERN FOUND

# Replace matching values with NaN
merged_zomato['Resturant_known_for'] = merged_zomato['Resturant_known_for'].replace(to_replace=pattern, value=np.nan, regex=True) # REPLACE THE PATTERN WITH 'nan'
```

```python
merged_zomato['Resturant_known_for'].value_counts() # CHECK IF SUCCESFULL
```

```python
merged_zomato # CHECK FOR DISCREPANCIES
```

```python
merged_zomato.drop_duplicates(subset='restaurant_name', keep='first',inplace=True) # REMOVE THE DUPLICATE HOTEL NAMES BUT KEEP THE FIRST VALUE
```

```python
merged_zomato # FINAL CHECK
```

```python
merged_zomato.to_csv("final_zomato.csv") # CONVERT THE CLEAN DATA TO CSV FILE FOR FURTHER ANALYSIS
```
